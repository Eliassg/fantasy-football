```python
from cognite.client import CogniteClient, ClientConfig
from dotenv import load_dotenv
import os
import pandas as pd

load_dotenv()
```

```python
def get_client() -> CogniteClient:
    config = {
            "client_name": "elias-test-streamlit-app-app",
            "project": os.getenv("CDF_PROJECT"),
            "base_url": os.getenv("CDF_BASE_URL"),
            "credentials": {
                "client_credentials": {
                    "client_id": os.getenv("IDP_CLIENT_ID"),
                    "client_secret": os.getenv("IDP_CLIENT_SECRET"),
                    "token_url": os.getenv("IDP_TOKEN_URL"),
                    "scopes": [f"{os.getenv('CDF_BASE_URL')}/.default"],
                },
            },
        }

    return CogniteClient.load(config)
```

```python
client = get_client()
```

```python
containers = client.data_modeling.containers.list(space="fantasy_football")
```

```python
client.data_modeling.containers.delete([(c.space, c.external_id) for c in containers])
```

```python
views = client.data_modeling.views.list(space="fantasy_football")
client.data_modeling.views.delete([(v.space, v.external_id, v.version) for v in views])
```

```python
from cognite.client.data_classes.data_modeling import ContainerApply

containers_path = ""

container_paths = []
for container_file in os.listdir(containers_path):
    container_paths.append(os.path.join(containers_path, container_file))

containers = []

for container_path in container_paths:
    with open(container_path, "r") as f:
        container_definition = f.read()
        containers.append(ContainerApply.load(container_definition))

client.data_modeling.containers.apply(containers)
        
```

```python
from cognite.client.data_classes.data_modeling import ViewApply

views_path = ""

view_paths = []
for view_file in os.listdir(views_path):
    view_paths.append(os.path.join(views_path, view_file))
views = []
for view_path in view_paths:
    with open(view_path, "r") as f:
        view_definition = f.read()
        views.append(ViewApply.load(view_definition))
client.data_modeling.views.apply(views)
```

```python
client.data_modeling.data_models.delete(("fantasy_football", "FantasyFootball", "1"))
```

```python
from cognite.client.data_classes.data_modeling import DataModelApply

data_model_path = ""

with open(data_model_path, "r") as f:
    data_model_definition = f.read()
    data_model = DataModelApply.load(data_model_definition)
    data_model.as_write()
    client.data_modeling.data_models.apply(data_model)
```
